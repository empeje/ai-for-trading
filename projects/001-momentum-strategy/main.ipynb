## What Does "Trading with Momentum" Mean?

Trading with momentum refers to an investment strategy that seeks to capitalize on the continuance of existing trends in the market. The core idea is that assets which have shown a strong price movement in a particular direction (either upward or downward) are likely to continue moving in that direction for some time.

Momentum traders typically look for stocks or other assets that are exhibiting high returns over a recent period, under the assumption that these trends will persist due to factors like investor psychology, herd behavior, or delayed information flow. They may use technical indicators such as moving averages, relative strength index (RSI), or volume spikes to identify and confirm momentum.

In summary, momentum trading is about "buying high and selling higher" (in an uptrend) or "selling low and buying lower" (in a downtrend), rather than trying to buy at the absolute bottom or sell at the absolute top.


## What Does the Formula $(p_t - p_{t-1}) / p_{t-1}$ Represent?

This formula computes the **simple return** (or percentage change) of an asset's price between two consecutive time points.

- **$p_t$**: Asset price at the current time step ($t$)
- **$p_{t-1}$**: Asset price at the previous time step ($t-1$)

The formula:

$$
\text{Return} = \frac{p_t - p_{t-1}}{p_{t-1}}
$$

This expresses the price change as a fraction of the previous price. For example, if a stock was \$100 yesterday and \$105 today:

$$
\frac{105 - 100}{100} = 0.05 \text{ or } 5\%
$$

So, the price increased by 5% over the period. This calculation is a core building block in finance for measuring performance, comparing assets, and designing trading strategies.


## Example: Simple Interest Calculation After 1 Year

Suppose you invest **$100** in an asset, and the rate of return is **4% per year**. The interest is only calculated and added after the money has been in the account for a full year (i.e., no compounding within the year).

**How much money do you have after 1 year?**

**Calculation:**

- Initial investment: $100
- Annual rate of return: 4% (as a decimal: 0.04)
- Interest accrued after 1 year: $100 × 0.04 = $4
- Total after 1 year: $100 + $4 = **$104**

**Answer:**  
After 1 year, you will have **$104** in your account.



```python
# Example: Simple Interest Calculation After 1 Year

initial_investment = 100
annual_rate = 0.04  # 4% as a decimal

interest = initial_investment * annual_rate
total_after_1_year = initial_investment + interest

print(f"Initial investment: ${initial_investment}")
print(f"Annual rate of return: {annual_rate * 100:.2f}%")
print(f"Interest accrued after 1 year: ${interest:.2f}")
print(f"Total after 1 year: ${total_after_1_year:.2f}")
```

## What is Compounding?

In finance, **compounding** means you earn interest not just on your original investment, but also on the interest that has been added over time. Each period, your interest is calculated on a growing balance.

**Key idea:**  
- With *simple interest*, you only earn interest on your initial investment.
- With *compound interest*, you earn interest on both your initial investment *and* any interest that has already been added.

**Example:**  
Suppose you invest $100 at a 4% annual rate, and the interest is **compounded annually**:

- After 1 year:  
  $100 × 1.04 = $104
- After 2 years:  
  $104 × 1.04 = $108.16
- After 3 years:  
  $108.16 × 1.04 ≈ $112.49

Notice how each year, the interest is calculated on a larger amount. Over time, compounding can significantly increase your returns compared to simple interest.

**Formula for compound interest after _n_ periods:**  
$$
\text{Final Amount} = \text{Initial Investment} \times (1 + r)^n
$$
where $r$ is the interest rate per period, and $n$ is the number of periods.




## How Often You Compound: Why It Matters

The **frequency of compounding**—how often interest is added to your investment—can make a noticeable difference in your returns over time.

### Common Compounding Periods
- **Annually:** Once per year
- **Semiannually:** Twice per year
- **Quarterly:** Four times per year
- **Monthly:** Twelve times per year
- **Weekly:** 52 times per year
- **Daily:** 365 times per year (sometimes 360)

### Why Does Compounding Frequency Matter?
Each time interest is compounded, the new interest is calculated on a slightly larger balance. The more frequently this happens, the more "interest on interest" you earn.

**Example:**  
Suppose you invest $1,000 at a 5% annual rate for 1 year.

- **Annually:**  
  $1,000 × (1 + 0.05) = $1,050.00  
- **Quarterly:**  
  $1,000 × (1 + 0.05/4)<sup>4</sup> ≈ $1,050.95  
- **Monthly:**  
  $1,000 × (1 + 0.05/12)<sup>12</sup> ≈ $1,051.16  
- **Daily:**  
  $1,000 × (1 + 0.05/365)<sup>365</sup> ≈ $1,051.27  

The difference is small over one year, but it grows with higher rates and longer time periods.

### General Formula for Compound Interest with _k_ Compounding Periods per Year

$$
\text{Final Amount} = \text{Initial Investment} \times \left(1 + \frac{r}{k}\right)^{k \cdot n}
$$

- $r$ = annual interest rate (as a decimal)
- $k$ = number of compounding periods per year
- $n$ = number of years

### Key Takeaway
The more frequently you compound, the more you earn—but the effect is most dramatic over long periods or at high rates. For most practical purposes, compounding daily or monthly yields only a slightly higher return than compounding annually, but over decades, the difference can add up!



```python
import numpy as np
import pandas as pd

# Parameters
principal = 1000.0
annual_rate = 0.05
years = 1

# Compounding frequencies and their periods per year
frequencies = {
    "Annually": 1,
    "Semiannually": 2,
    "Quarterly": 4,
    "Monthly": 12,
    "Weekly": 52,
    "Daily": 365,
}

results = []

for label, k in frequencies.items():
    amount = principal * (1 + annual_rate / k) ** (k * years)
    results.append({"Frequency": label, "Periods/Year": k, "Final Amount": amount})

# Continuous compounding
continuous_amount = principal * np.exp(annual_rate * years)
results.append({"Frequency": "Continuous", "Periods/Year": "∞", "Final Amount": continuous_amount})

df = pd.DataFrame(results)
df["Final Amount"] = df["Final Amount"].map(lambda x: f"${x:,.2f}")

display(df)

# Optionally, visualize convergence
import matplotlib.pyplot as plt

k_values = np.logspace(0, 5, 100)  # from 1 to 100,000 periods/year
amounts = principal * (1 + annual_rate / k_values) ** (k_values * years)
continuous = principal * np.exp(annual_rate * years)

plt.figure(figsize=(8, 5))
plt.plot(k_values, amounts, label="Discrete Compounding")
plt.axhline(continuous, color='r', linestyle='--', label="Continuous Compounding")
plt.xscale('log')
plt.xlabel("Compounding Periods per Year (log scale)")
plt.ylabel("Final Amount ($)")
plt.title("Convergence of Compounding to Continuous Limit")
plt.legend()
plt.grid(True, which="both", ls="--", alpha=0.5)
plt.show()
```

```python
# Simulate the difference in final amount after 10 years for each compounding frequency, and show net interest

years_10 = 10
results_10y = []

for label, k in frequencies.items():
    amount_10y = principal * (1 + annual_rate / k) ** (k * years_10)
    net_interest_10y = amount_10y - principal
    results_10y.append({
        "Frequency": label,
        "Periods/Year": k,
        "Final Amount": amount_10y,
        "Net Interest": net_interest_10y
    })

# Continuous compounding for 10 years
continuous_amount_10y = principal * np.exp(annual_rate * years_10)
continuous_net_interest_10y = continuous_amount_10y - principal
results_10y.append({
    "Frequency": "Continuous",
    "Periods/Year": "∞",
    "Final Amount": continuous_amount_10y,
    "Net Interest": continuous_net_interest_10y
})

df_10y = pd.DataFrame(results_10y)
df_10y["Final Amount"] = df_10y["Final Amount"].map(lambda x: f"${x:,.2f}")
df_10y["Net Interest"] = df_10y["Net Interest"].map(lambda x: f"${x:,.2f}")

print("Final Amount after 10 years for each compounding frequency:")
display(df_10y)

# Visualize the difference after 10 years
k_values_10y = np.logspace(0, 5, 100)
amounts_10y = principal * (1 + annual_rate / k_values_10y) ** (k_values_10y * years_10)
continuous_10y = principal * np.exp(annual_rate * years_10)

plt.figure(figsize=(8, 5))
plt.plot(k_values_10y, amounts_10y, label="Discrete Compounding (10 years)")
plt.axhline(continuous_10y, color='r', linestyle='--', label="Continuous Compounding (10 years)")
plt.xscale('log')
plt.xlabel("Compounding Periods per Year (log scale)")
plt.ylabel("Final Amount ($)")
plt.title("Convergence of Compounding to Continuous Limit (10 Years)")
plt.legend()
plt.grid(True, which="both", ls="--", alpha=0.5)
plt.show()
```

### Given: final amount after continuous compounding, principal, and years

Let:

- $A$ = final amount after compounding  
- $P$ = principal (initial amount)  
- $r$ = continuously compounded annual rate  
- $t$ = number of years  

The formula for continuous compounding is:

$$
A = P \cdot e^{r t}
$$

To solve for $r$:

$$
r = \frac{\ln\left(\frac{A}{P}\right)}{t}
$$

**Example:**  
If $A = 104.08$, $P = 100$, $t = 1$,

$$
r = \frac{\ln\left(\frac{104.08}{100}\right)}{1} \approx 0.04 = 4\%
$$



```python
# Given: final amount after continuous compounding, principal, and years
final_amount = 104.08
principal = 100
years = 1

# Solve for r in: final_amount = principal * exp(r * years)
# => r = ln(final_amount / principal) / years
r = np.log(final_amount / principal) / years

print(f"The implied continuously compounded rate r is: {r:.4%}")
```

### Additivity Principle in Continuously Compounded Rates

The additivity principle states that with continuous compounding, the total effect over multiple periods is simply the sum of the individual rates for each period.

Suppose you have a continuously compounded rate of 2% per month. To find the total effect after 12 months, add the rates for each month:

$$
\text{Total rate over 12 months} = 0.02 \times 12 = 0.24 \quad \text{(or 24\%)}
$$

The formula for continuous compounding is:

$$
A = P \cdot e^{r_{\text{total}}}
$$

where $r_{\text{total}}$ is the sum of the continuously compounded rates over all periods.

If you invest $P$ at a 2% continuously compounded rate per month for 12 months, the final amount is:

$$
A = P \cdot e^{0.02 \times 12} = P \cdot e^{0.24}
$$

This additivity means you can simply add the rates for each period, rather than multiplying factors as you would with discrete compounding.

For example, the effect after each month is:

- After 1 month: $A_1 = P \cdot e^{0.02}$
- After 2 months: $A_2 = P \cdot e^{0.02 + 0.02} = P \cdot e^{0.04}$
- ...
- After 12 months: $A_{12} = P \cdot e^{0.02 \times 12} = P \cdot e^{0.24}$

This property makes calculations with continuously compounded rates especially convenient, as the total exponent is just the sum of the individual rates over time.


### Annualized Rate of Return and the Principle of Additivity

When using continuously compounded rates, the principle of additivity allows us to easily annualize a monthly rate. If the monthly continuously compounded rate is $r_{monthly}$, then the annualized rate is simply $r_{annual} = r_{monthly} \times 12$. This is because, with continuous compounding, the total rate over multiple periods is just the sum of the individual rates.

For example, if $r_{monthly} = 0.02$ (2%), then the annualized rate is $0.02 \times 12 = 0.24$ (or 24%). The final amount after one year would be $A = P \cdot e^{r_{annual}} = P \cdot e^{0.24}$.

This linear relationship makes it straightforward to convert between different compounding periods when working with continuously compounded returns.

### Time Additivity of Log Returns

Log returns, or continuously compounded returns, are additive over time. If $P_0$ is the initial price and $P_1, P_2, \ldots, P_n$ are the prices at subsequent times, the log return for each period is

$$
r_i = \ln\left(\frac{P_i}{P_{i-1}}\right)
$$

The total log return over $n$ periods is the sum of the individual log returns:

$$
R_{\text{total}} = \sum_{i=1}^n r_i = \ln\left(\frac{P_1}{P_0}\right) + \ln\left(\frac{P_2}{P_1}\right) + \cdots + \ln\left(\frac{P_n}{P_{n-1}}\right)
$$

By the properties of logarithms, this sum simplifies to

$$
R_{\text{total}} = \ln\left(\frac{P_n}{P_0}\right)
$$

This demonstrates that log returns are time-additive: the total log return over multiple periods is simply the log of the ratio of the final price to the initial price.



```python
import numpy as np
import pandas as pd

# Example: Time Additivity of Log Returns (January and February)
prices = pd.Series(
    [100, 110, 121],  # P0, P1, P2
    index=["Jan 1", "Feb 1", "Mar 1"]
)

# Calculate log returns for each month
log_return_jan = np.log(prices["Feb 1"] / prices["Jan 1"])
log_return_feb = np.log(prices["Mar 1"] / prices["Feb 1"])

# Total log return over two months (sum of monthly log returns)
total_log_return_sum = log_return_jan + log_return_feb

# Direct log return from start to end
total_log_return_direct = np.log(prices["Mar 1"] / prices["Jan 1"])

# Display results
print("Price Series:")
print(prices)
print("\nLog Return January:    {:.5f}".format(log_return_jan))
print("Log Return February:   {:.5f}".format(log_return_feb))
print("Sum of Log Returns:    {:.5f}".format(total_log_return_sum))
print("Direct Log Return:     {:.5f}".format(total_log_return_direct))
print("\nAre the total log returns equal? {}".format(
    np.isclose(total_log_return_sum, total_log_return_direct)
))

# For clarity, show the percent returns as well
jan_pct = (prices["Feb 1"] / prices["Jan 1"] - 1) * 100
feb_pct = (prices["Mar 1"] / prices["Feb 1"] - 1) * 100
print("\nPercent Return January:  {:.2f}%".format(jan_pct))
print("Percent Return February: {:.2f}%".format(feb_pct))
```

### Numerical Stability of Logarithms in Return Calculations

When calculating log returns, numerical stability is generally reliable as long as all price values are strictly positive. The logarithm function is only defined for positive numbers; if a price is zero or negative (due to data errors or missing values), the result will be NaN or infinite, which can silently propagate through your calculations.

For very small price changes, floating-point precision can become an issue. In particular, when computing log(1 + x) for small x (such as small percent returns), using np.log1p(x) is recommended. This function is numerically stable for values of x near zero and avoids loss of precision that can occur with np.log(1 + x).

Best practices:
- Always validate that all prices are positive before taking logarithms.
- Use np.log1p(x) instead of np.log(1 + x) for small returns.
- Be aware that extremely large or small values can still cause floating-point overflows or underflows, though this is rare in typical financial data.

Example:
small_return = 1e-8
log_return = np.log1p(small_return)  # More accurate than np.log(1 + small_return)

In summary, log returns are robust for most financial data, but careful data validation and numerically stable functions help prevent subtle bugs.



```python
import numpy as np

arr = np.arange(1, 1000)
reciprocals = 1 / arr

# Direct product: for large arrays, this can underflow to zero (all precision lost)
prod_reciprocals = np.prod(reciprocals)

# Logarithmic sum: numerically stable, avoids underflow
sum_log_reciprocals = np.sum(np.log(reciprocals))

# Exponentiate the sum of logs to recover the product in a stable way
exp_sum_log = np.exp(sum_log_reciprocals)

are_close = np.isclose(prod_reciprocals, exp_sum_log)

print("Direct product of reciprocals (prone to underflow):", prod_reciprocals)
print("Sum of logs of reciprocals (stable):", sum_log_reciprocals)
print("Exp(sum of logs) (stable product):", exp_sum_log)
print("Are the direct product and exp(sum(logs)) close?", are_close)

# Why is this important?
# In financial and scientific computing, we often need to multiply many small numbers (e.g., cumulative returns, likelihoods).
# Direct multiplication can quickly underflow to zero, especially as the number of terms grows, causing total loss of information.
# By summing logarithms and exponentiating, we keep intermediate values in a numerically safe range, preserving accuracy.
# This log-sum-exp trick is a fundamental technique for robust numerical computation in any domain where products of many numbers arise.
```
